```python
#Import the necessary packages

import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

plt.style.use('science') #https://github.com/garrettj403/SciencePlots
```

```python
#Read the data from the .txt file as a 2d array

data = np.loadtxt('data.txt')

I_cs = data[:,0]
x_cs = data[:,1]

I_co1 = data[:,2]
x_co1 = data[:,3]

I_co2 = data[:,4]
x_co2 = data[:,5]

I_na = data[:,6]
x_na = data[:,7]

errors = x_na*0 #In order to use plt.errorbar()
```

```python
#Define the fit function
def fit (x, a, b):
	return a * np.exp(-x*b)

x = np.linspace(1, 9, 1000)
```

```python
popt, pcov = curve_fit(fit, x_cs, I_cs, p0=(1577, 1))
print("Cs-137 Fit results:")
print(popt)
print(np.sqrt(np.diag(pcov)))

plt.errorbar(x_cs, I_cs, errors, errors, 'bo')
plt.plot(x, fit(x, *popt), 'r--')
plt.grid()
plt.title('$I = f(\chi ), ^{137}Cs$')
plt.ylabel('$I (counts/sec)$')
plt.xlabel('$\chi ( gr/cm^2)$')
plt.legend(['Fit function', 'Experimental data'])
plt.savefig('I_cs.png', dpi=500)
```
